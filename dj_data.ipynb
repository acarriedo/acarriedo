```python
import pandas as pd
import plotly.express as px
```

5 - 8.

```python
# Read CSV for the column you need for lounger audience
''' 
LOUNGER / QUIET TYPES AUDIENCE
'''

df = pd.read_csv("music_genre_classification.csv", usecols
= ['Artist Name', 'energy', 'loudness', 'acousticness'])

# Get the smallest 10 energy values

low_energy = df.nsmallest (10, 'energy')
print(low_energy[['Artist Name', 'energy']])

# Plot a bar chart, using plotly

fig = px.bar(low_energy, 'Artist Name', 'energy', color='energy', color_continuous_scale='Greens')
fig.update_layout(title_text='10 Lowest Energy Songs by Artist', title_x=.5)

```

9 - 10.

```python


# Get the lowest 10 loudness values in decibels

quietest = df.nsmallest(10, 'loudness')
print(quietest[['Artist Name', 'loudness']])

# Plot to a bar chart in Plotly

fig = px.bar(quietest, 'Artist Name', 'loudness', color='loudness', color_continuous_scale='Blues')
fig.update_layout(title_text='10 Quietest Songs by Artist', title_x=.5)


```

11.

```python
# Get the highest 10 acousticness values

best_acoustics = df.nlargest(10, 'acousticness')
print(best_acoustics[['Artist Name', 'acousticness']])

# Plot to a bar chart in plotly

fig = px.bar(best_acoustics, 'Artist Name', 'acousticness', color ='acousticness', color_continuous_scale='sunsetdark')
fig.update_layout(title_text='10 Best Acoustic Sounds by Artist', title_x=.5)

```

```python
# Viewing all columns for Music Genre Classification

df = pd.read_csv("music_genre_classification.csv")
df.head(30)
```

Create visualizations for: Track Team / Partiers and the NYT Bestseller Book Club / Literary fans

```python
# Track Team Ordered by Tempo

track_tempo = df.nlargest(15, 'tempo')
print(track_tempo[['Track Name', 'tempo']])

fig = px.bar(track_tempo, 'Track Name', 'tempo', color ='tempo', color_continuous_scale='tempo')
fig.update_layout(title_text='15 Best Tempo Songs', title_x=.5)

```

```python
# Partiers Ordered by Danceability

best_for_dancing = df.nlargest(10, 'danceability')
print(best_for_dancing[['Track Name', 'danceability']])

fig = px.bar(best_for_dancing, 'Track Name', 'danceability', color ='danceability', color_continuous_scale='oranges')
fig.update_layout(title_text='10 Best Dance Songs', title_x=.5)
```

```python
# NYT Best Seller Book Club Ordered by Instrumentals

best_instrumentals = df.nlargest(10, 'instrumentalness')
print(best_instrumentals[['Track Name', 'instrumentalness']])

fig = px.bar(best_instrumentals, 'Track Name', 'instrumentalness', color ='instrumentalness', color_continuous_scale='deep')
fig.update_layout(title_text='10 Best Soft Instrumentals', title_x=.5)
```

```python
# NYT Best Seller Book Club Ordered by Low Speech

lyric_free = df.nsmallest(10, 'speechiness')
print(lyric_free[['Track Name', 'speechiness']])

fig = px.bar(lyric_free, 'Track Name', 'speechiness', color ='speechiness', color_continuous_scale='earth')
fig.update_layout(title_text='10 Best Book Club Songs', title_x=.5)
```
